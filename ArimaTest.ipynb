```python
merge_df = merge_df.iloc[50:]
```

```python
merge_df
```

```python
pm25 = merge_df[["일시", "PM-2.5"]]
pm25["일시"] = pd.to_datetime(pm25["일시"], infer_datetime_format = True)
pm25.info()
```

```python
pm25 = pm25.set_index("일시")
```

```python
pm25
```

```python
trn = pm25[:-200] # 2018-02-01 ~ 2023-02-12 / 1829
tst = pm25[-200:] # 2023-02-13 ~ 2023-08-31 / 200
```

```python
import matplotlib.pyplot as plt
# 훈련 데이터의 그래프 그리기 (파란색)
plt.plot(trn.index, trn['PM-2.5'], label='Training Data')

# 테스트 데이터의 그래프 그리기 (빨간색)
plt.plot(tst.index, tst['PM-2.5'], label='Test Data')
```

```python
from statsmodels.tsa.arima.model import ARIMA
arima_model = ARIMA(trn, order=(1,0,1))
arima_model_fit = arima_model.fit()
# print(model_fit.summary())
# start = pd.to_datetime("2023-02-13")
# end = pd.to_datetime("2023-08-31")
# dates = pd.date_range(start, end)
prd = arima_model_fit.predict("2023-02-13", "2023-08-31", dynamic=True)
prd = pd.DataFrame(prd)
prd
```

```python
plt.plot(tst, label="test")
plt.plot(prd, label="predict")
plt.legend()
```

```python
from pmdarima.arima import auto_arima

model_auto = auto_arima(tst, start_p=0, start_q=0, max_p=50, max_q=50, seasonal=True, m=12, trace=True)
```

```python
arima_model2 = ARIMA(trn, order=(0,1,0), seasonal_order=(0,0,0,365))
arima_model_fit2 = arima_model2.fit()
# print(arima_model_fit2.summary())
prd2 = arima_model_fit2.predict("2023-02-13", "2023-08-31", dynamic=True)
prd2 = pd.DataFrame(prd2)
prd2
```

```python
plt.plot(tst, label="test")
plt.plot(prd2, label="predict")
plt.legend()
```
