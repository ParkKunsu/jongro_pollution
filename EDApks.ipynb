# 사용 컬럼
+ 평균 해면기압 - 하루 동안 관측된 해면기압의 평균값  
+ 합계 일사량 - 일조시간 / 가조시간   
+ 평균 지면온도 - 일정조건에서 측정된 지면온도   
+ 최저 초상온도 - 지표에 접해 있는 풀 위의 공기온도, 서리 발생을 가늠하는 척도   
+ 평균 풍속 - 특정 기간 동안의 평균 풍속을 미터/초 단위로 나타낸 것   
+ 최다 풍향 - 특정 기간 동안 가장 자주 나타나는 풍향을 16방위로 표현한 것   
+ 평균 상대습도 - 특정 기간 동안의 평균 상대습도로, 공기에 포함된 수증기의 평균 양에 대한 백분율   
+ 일 최심적설 - 눈이 내린 후, 그 중에서 가장 두꺼운 층의 두께   
+ 평균 전운량  - 하루동안 구름 양을 나타내는 지표 10으로 갈수록 구름이 많아짐  
+ 평균기온(°C)  
+ 최저기온(°C)  
+ 최고기온(°C)  
+ 강수 계속시간(hr)  
+ 일강수량(mm)

# 시각화

### 이상치 분석

```python
import pandas as pd
data = pd.read_csv("/home/dataset/complete.csv")

trn = data[["일시", "PM-2.5"]]
trn["일시"] = pd.to_datetime(trn["일시"], infer_datetime_format = True)
trn = trn.set_index("일시")
trn = trn.iloc[50:]
```

```python
import matplotlib.pyplot as plt
# import seaborn as sns

# 데이터 프레임의 열을 선택하여 히스토그램 그리기
# plt.hist(data['PM-2.5'], bins=30, edgecolor='black')
# plt.scatter(data['PM-2.5'], data['PM-10'])
plt.boxplot(x=data['PM-2.5'])

plt.show()
```

### ACF, PACF
+ 그래프상 AR모델 적용하기 좋을것이라 생각

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(211)
fig = plot_acf(trn.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(trn.values.squeeze(), lags=40, ax=ax2)
```

### 데이터 수치확인

```python
# p-value 4.213611925296449e-10로 충분히 낮음
from statsmodels.tsa.stattools import adfuller
adf = adfuller(trn)
print('ADF Statistic: {}'.format(adf[0]))
print('p-value: {}'.format(adf[1]))
print('Critical Values:')
for key, value in adf[4].items():
  print('\t{}: {}'.format(key, value))
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
trn_diff = trn.diff(365).dropna()
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(211)
fig = plot_acf(trn_diff.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(trn_diff.squeeze(), lags=40, ax=ax2)

adf = adfuller(trn_diff)
print('ADF Statistic: {}'.format(adf[0]))
print('p-value: {}'.format(adf[1]))
print('Critical Values:')
for key, value in adf[4].items():
  print('\t{}: {}'.format(key, value))
trn_diff = trn.diff(365).dropna()
```

```python
trn_diff = trn - trn.shift()
plt.plot(trn_diff)
```

```python
# 차분시 p-value 더 낮아지긴함
from statsmodels.tsa.stattools import adfuller
adf = adfuller(trn_diff.dropna())
print('ADF Statistic: {}'.format(adf[0]))
print('p-value: {}'.format(adf[1]))
print('Critical Values:')
for key, value in adf[4].items():
  print('\t{}: {}'.format(key, value))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decom = seasonal_decompose(trn)

trend = decom.trend
seasonal = decom.seasonal
residual = decom.resid

fig = plt.figure(figsize=(30, 20))

plt.subplot(4, 1, 1)
plt.plot(trn, label= "original")
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(trend, label= "trend")
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(seasonal, label= "seasonal")
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(residual, label= "residual")
plt.legend()
```
